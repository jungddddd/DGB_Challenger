```python
#Î∂ÄÎèôÏÇ∞ Ïπ¥ÌÖåÍ≥†Î¶¨ TF-IDF

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv("Desktop/output/Î∂ÄÎèôÏÇ∞.csv")

# ÌÖçÏä§Ìä∏ Î¶¨Ïä§Ìä∏ ÏÉùÏÑ± (Í≤∞Ï∏°Ïπò Ï†úÍ±∞)
texts = df['filtered_tokenized'].fillna("").astype(str).tolist()

# TF-IDF Î≤°ÌÑ∞Ìôî
vectorizer = TfidfVectorizer(
    max_df=0.95,
    min_df=2,
    ngram_range=(1, 2),
    max_features=1000
)
tfidf_matrix = vectorizer.fit_transform(texts)

# Ï†ÑÏ≤¥ ÏΩîÌçºÏä§ Í∏∞Ï§Ä TF-IDF Ï†êÏàò ÌèâÍ∑† Í≥ÑÏÇ∞
mean_tfidf_scores = tfidf_matrix.mean(axis=0).A1  
feature_names = vectorizer.get_feature_names_out()

# ÏÉÅÏúÑ 20Í∞ú ÌÇ§ÏõåÎìú Ï§ë Ï†úÏô∏Ìï† Îã®Ïñ¥ Ï†úÍ±∞
exclude_words = ['Î∂ÄÎèôÏÇ∞', 'Ìï©ÎãàÎã§', 'Ìï®Íªò','ÏïàÎÖï','Î∞îÎ°ú','ÏõêÌïò','ÏÉùÍ∞Å','ÏßÄÍ∏à','ÌñâÎ≥µ']
top_n = 20

keyword_scores = [(feature_names[i], mean_tfidf_scores[i]) for i in mean_tfidf_scores.argsort()[::-1]]
filtered_keywords = [(word, score) for word, score in keyword_scores if word not in exclude_words]

# Í≤∞Í≥º Ï∂úÎ†•
print("üìå Î∂ÄÎèôÏÇ∞ ÏÉÅÏúÑ TF-IDF ÌÇ§ÏõåÎìú :")
for word, score in filtered_keywords[:top_n]:
    print(f"{word}: {score:.4f}")
```
